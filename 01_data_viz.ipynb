# Count Plot - for counting categories

```python
sns.countplot(df['gender'],hue=df['marital_status'])
```

### Interpretation:
Tip (More broad info always come first then _**no**_ statement)
- Females are quite less than when compared to the males; whats the ratio?
-  Married ratio of females are greater than the males.
-  There are more single males then to the married males.


```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(data=df, x='gender', hue='favourite_dish')
# plt.xticks(size=12)
# plt.xlabel('Gender', size=14)
# plt.yticks(size=12)
# plt.ylabel('Count', size=12)
#plt.title("Fav dishes wrt gender", size=16)
# ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%\n'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center')
#plt.tight_layout()
plt.show()
```

## Interpretation
- Biryani is the top fav dish in the male section whereas,
- Naan haleem + saag is the top fav dish among females.
- Karahi gost is at the 2nd position among both groups.
- DAAL is the least fav among females wheras naan haleem is least fav among males.

# Outliers with IQR

```python
''' Detection '''
# IQR
Q1 = np.percentile(df['height'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(df['height'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['height'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['height'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)
```
